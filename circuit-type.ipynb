```
circuits = {
    "AUS": "M",  # Australian GP
    "CHN": "M",  # Chinese GP
    "JPN": "M",  # Japanese GP
    "MIA": "M",  # Miami GP
    "CAN": "M",  # Canadian GP
    "MON": "L",  # Monaco GP
    "ESP": "M",  # Spanish GP
    "AUT": "H",  # Austrian GP
    "GBR": "H",  # British GP
    "BEL": "H",  # Belgian GP
    "HUN": "L",  # Hungarian GP
    "NED": "M",  # Dutch GP
    "ITA": "H",  # Italian GP
    "MAD": "M",  # Madrid Street Circuit
    "AZE": "H",  # Azerbaijan GP
    "SGP": "L",  # Singapore GP
    "USA": "M",  # United States GP
    "MEX": "H",  # Mexico City GP
    "BRA": "M",  # Sao Paulo GP
    "LVG": "H",  # Las Vegas GP
    "QAT": "M",  # Qatar GP
    "ABU": "M"   # Abu Dhabi GP
}
```

```
import matplotlib.pyplot as plt
import itertools

driver = input('Driver: ')
num_races = int(input("Number of races: "))
races = dict(itertools.islice(circuits.items(), num_races))
low = []
medium = []
high = []
for race, circuit_type in races.items():
  print(f"Grand prix: {race}, circuit type: {circuit_type} speed")
  if circuit_type == "L":
    points = int(input('Fantasy points: '))
    low.append(points)
  elif circuit_type == "M":
    points = int(input('Fantasy points: '))
    medium.append(points)
  elif circuit_type == "H":
    points = int(input('Fantasy points: '))
    high.append(points)

x_axis = [f'Low (races: {len(low)})', f'Medium (races: {len(medium)})', f'High (races: {len(high)})']
y_axis = [sum(low), sum(medium), sum(high)]
plt.bar(x_axis, y_axis)
plt.title(driver)
plt.xlabel('Circuit Type (speed)')
plt.ylabel('Fantasy Points')
plt.show()
```

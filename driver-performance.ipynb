```
import matplotlib.pyplot as plt

points = input('Fantasy points progression: ').split() # separate using spaces
data = list(map(lambda x: int(x), points))

x = [n for n in range(1, len(data) + 1)]
y = [n for n in data]
plt.plot(x, y, marker=".", markersize=10, color='blue')
plt.xlabel('Race Number')
plt.ylabel('Fantasy Points')
plt.show()
```

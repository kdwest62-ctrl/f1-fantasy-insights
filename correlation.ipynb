```python
import matplotlib.pyplot as plt

constructor = input('Constructor: ').upper()
constructor_data = constructor_pts[constructor]

driver1 = input('Driver 1: ').upper()
driver1_data = driver_pts[driver1]
color1 = input('Marker color: ')

driver2 = input('Driver 2: ').upper()
driver2_data = driver_pts[driver2]
color2 = input('Marker color: ')

fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.scatter(constructor_data, driver1_data, color=color1, label=driver1)
ax1.set_xlabel(f'{constructor} Points')
ax1.set_ylabel(f'{driver1} Points')
ax1.tick_params(axis='y')

ax2 = ax1.twinx()
ax2.scatter(constructor_data, driver2_data, color=color2, label=driver2)
ax2.set_ylabel(f'{driver2} Points')
ax2.tick_params(axis='y')

plt.title(f'{constructor} Points vs. {driver1} and {driver2} Points Progression')
ax1.grid(True)

fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))
plt.show()
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

def stand_dev(nums):
  avg = np.mean(nums)
  sqrd_diff = []
  for num in nums:
    sqrd_diff.append((avg - num)**2)
  result = math.sqrt(sum(sqrd_diff)/len(sqrd_diff))
  return round(result, 2)

driver1 = input('Driver 1: ').upper()
driver1_pts = np.array(driver_pts[driver1])
result1 = stand_dev(driver1_pts)

driver2 = input('Driver 2: ').upper()
driver2_pts = np.array(driver_pts[driver2])
result2 = stand_dev(driver2_pts)

x_axis = [driver1, driver2]
y_axis = [result1, result2]

plt.bar(x_axis, y_axis)
plt.xlabel('Driver')
plt.ylabel('Standard Deviation')
plt.show()
```

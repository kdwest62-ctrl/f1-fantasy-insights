```python
import matplotlib.pyplot as plt

points = input('Fantasy points progression: ').split() # separated by spaces
data = list(map(lambda x: int(x), points))

x = [n for n in range(1, len(data) + 1)]
y = [n for n in data]
plt.plot(x, y, marker=".", markersize=10, color='blue')
plt.show()
```

```python
values = []
updates = []
for num in range(1, 5 + 1):
  value = float(input(f"Driver {num} current value (M): "))
  values.append(value)
  update = float(input("Value update (M): "))
  updates.append(update)
for num in range(1, 2 + 1):
  value = float(input(f"Constructor {num} current value (M): "))
  values.append(value)
  update = float(input("Value update (M): "))
  updates.append(update)

team_value = sum(values)
total_updates = sum(updates)
print("-" * 8)
print(f"Current team value: {team_value}")
print(f"Total updates: {total_updates}")
print(f"New team value: {team_value + total_updates}")
```

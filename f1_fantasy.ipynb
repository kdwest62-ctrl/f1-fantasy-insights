```python
import matplotlib.pyplot as plt

points = input('Fantasy points progression: ').split() # separated by spaces
data = list(map(lambda x: int(x), points))

x = [n for n in range(1, len(data) + 1)]
y = [n for n in data]
plt.plot(x, y, marker=".", markersize=10, color='blue')
plt.show()
```

```python
def value_update(value, update):
  if value >= 18.5:
    if update == "+":
      return 0.3
    elif update == "-":
      return -0.3
  elif value < 18.5:
    if update == "+":
      return 0.6
    elif update == "-":
      return -0.6
values = []
updates = []
for num in range(1, 5 + 1):
  current_val = float(input(f"Driver {num} current value (M): "))
  values.append(current_val)
  update_choice = input("Value update (+ or -): ")
  updates.append(value_update(current_val, update_choice))
for num in range(1, 2 + 1):
  current_val = float(input(f"Constructor {num} current value (M): "))
  values.append(current_val)
  update_choice = input("Value update (+ or -): ")
  updates.append(value_update(current_val, update_choice))
cost_cap = float(input("Cost cap (M): "))
team_value = sum(values)
total_updates = sum(updates)
print("-" * 8)
print(f"Value of selected team: {team_value}")
print(f"Value of selected team + cost cap: {team_value + cost_cap}")
print(f"Total updates: {round(total_updates, 1)}")
print(f"Updated value of selected team: {team_value + total_updates}")
print(f"Updated value of selected team + cost cap: {team_value + total_updates + cost_cap}")
```

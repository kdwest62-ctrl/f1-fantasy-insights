```python
def value_update(value, update):
  if value >= 18.5:
    if update == "+":
      return 0.3
    elif update == "-":
      return -0.3
  elif value < 18.5:
    if update == "+":
      return 0.6
    elif update == "-":
      return -0.6
values = []
updates = []
for num in range(1, 5 + 1):
  driver = input(f"Driver {num}: ").upper()
  values.append(driver_val[driver])
  update_choice = input("Value update (+ or -): ")
  updates.append(value_update(driver_val[driver], update_choice))
for num in range(1, 2 + 1):
  constructor = input(f"Constructor {num}: ").upper()
  values.append(constructor_val[constructor])
  update_choice = input("Value update (+ or -): ")
  updates.append(value_update(constructor_val[constructor], update_choice))
cost_cap = float(input("Cost cap (M): "))
team_value = sum(values)
total_updates = sum(updates)
print("-" * 8)
print(f"Value of selected team: {team_value}")
print(f"Value of selected team + cost cap: {round(team_value + cost_cap, 1)}")
print(f"Total updates: {round(total_updates, 1)}")
print(f"Updated value of selected team: {round(team_value + total_updates, 1)}")
print(f"Updated value of selected team + cost cap: {round(team_value + total_updates + cost_cap, 1)}")
```

```python
driver = input("Driver: ").upper()
value = driver_val[driver]
if value < 18.5:
  print(f"Current value (M): {value}")
  upgrade = float(input("Select constant upgrade (0.6 or 0.2): "))
  races = 0
  while value < 18.5:
    value += upgrade
    races += 1
  print(f"Races needed to get to Tier A: {races}")
  races_remain = 22 - len(driver_pts[driver])
  print(f"Races remaining: {races_remain}")
elif value >= 18.5:
  print(f"Current value: {value}")
  downgrade = float(input("Select constant downgrade (0.3 or 0.1): "))
  races = 0
  while value >= 18.5:
    value -= downgrade
    races += 1
  print(f"Races needed to go down to Tier B: {races}")
  races_remain = 22 - len(driver_pts[driver])
  print(f"Races remaining: {races_remain}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

driver = input("Driver: ").upper()
data = np.array(driver_pts[driver])
avg = np.mean(data)

over = 0
equal = 0
under = 0
for num in data:
  if num > avg:
    over += 1
  elif num == avg:
    equal += 1
  elif num < avg:
    under += 1
over_pct = (over/len(data)) * 100
equal_pct = (equal/len(data)) * 100
under_pct = (under/len(data)) * 100

categories = ['Over AVG', 'Equal AVG', 'Under AVG']
values = [over_pct, equal_pct, under_pct]
fig, ax = plt.subplots()
bars = ax.bar(categories, values, color='skyblue')
ax.bar_label(bars, label_type='center', color='white', fontsize=12, weight='bold')
plt.ylim(0, 100)
plt.title(f'Average Fantasy Points ({driver}): {avg}')
plt.xlabel('Tendency to Score')
plt.ylabel('Percentage (%)')
plt.show()
```

```python
import numpy as np

def mean(nums):
  avg = np.mean(nums)
  return round(avg, 2)

points = input('Fantasy points progression: ').split() # separate using spaces
data = np.array(list(map(lambda x: int(x), points)))
print(f"Average points: {mean(data)}")
```

```python
import math
import numpy as np

def stand_dev(nums):
  avg = np.mean(nums)
  sqrd_diff = []
  for num in nums:
    sqrd_diff.append((avg - num)**2)
  result = math.sqrt(sum(sqrd_diff)/len(sqrd_diff))
  return round(result, 2)

points = input('Fantasy points progression: ').split() # separate using spaces
data = np.array(list(map(lambda x: int(x), points)))
print(f"Standard deviation: {stand_dev(data)}")
```

Let's break down the standard deviation formula step-by-step:

1.  **Calculate the Mean ($\bar{x}$):** Sum all the data points ($x_i$) and divide by the number of data points ($N$).

    $$\bar{x} = \frac{\sum x_i}{N}$$

2.  **Calculate the Deviation from the Mean:** For each data point ($x_i$), subtract the mean ($\bar{x}$). This tells you how far each point is from the average.

    $$(x_i - \bar{x})$$

3.  **Square the Deviations:** Square each of the deviations calculated in the previous step. This is done to eliminate negative values and to give more weight to larger deviations.

    $$(x_i - \bar{x})^2$$

4.  **Sum the Squared Deviations:** Add up all the squared deviations.

    $$\sum (x_i - \bar{x})^2$$

5.  **Calculate the Variance (for a sample):** Divide the sum of the squared deviations by ($N-1$). We use ($N-1$) for a sample to get an unbiased estimate of the population variance.

    $$\text{Variance} = \frac{\sum (x_i - \bar{x})^2}{N-1}$$

6.  **Take the Square Root (Standard Deviation):** Finally, take the square root of the variance to get the standard deviation ($\sigma$). This brings the value back to the original units of the data.

    $$\sigma = \sqrt{\frac{\sum (x_i - \bar{x})^2}{N-1}}$$
